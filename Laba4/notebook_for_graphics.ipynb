```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из входного файла .csv
linear_data = pd.read_csv('time_file_linear.csv', sep=',', header=None, names=['n', 'time'])
log_data = pd.read_csv('time_file_nlogn.csv', sep=',', header=None, names=['n', 'time'])

# Построение графика
plt.plot(linear_data['n'], linear_data['time'], '-', log_data['n'], log_data['time'], '--')
plt.xlabel('Sizes of arrays')
plt.ylabel('Time taken (microsec)')
plt.title('Compare the implementations of HeapBuild')
plt.legend(['Build heap for O(N)', 'Build heap for O(NlogN)'])
plt.savefig('graphic.png')
plt.show()
```
