```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из входного файла .txt
data1 = pd.read_csv('TimeTests/insert_avl_file.csv', sep=',', header=None, names=['n', 'time'])
data2 = pd.read_csv('TimeTests/delete_avl_file.csv', sep=',', header=None, names=['n', 'time'])

# Построение графика
plt.plot(data1['n'], data1['time'], '-')
plt.xlabel('Number of elements')
plt.ylabel('Time taken (milliseconds)')
plt.legend('Insert in AVL-tree')
plt.title('Graphic of insertion time(number of elements)')
plt.savefig('Images/avl_tree_insert.png')
plt.show()
```
