```python
import pandas as pd
import matplotlib.pyplot as plt

#      .txt
data1 = pd.read_csv('TimeFiles/bubble_sort_time.csv', sep=';', header=None, names=['n', 'time'])
data2 = pd.read_csv('TimeFiles/insertion_sort_time.csv', sep=';', header=None, names=['n', 'time'])
data3 = pd.read_csv('TimeFiles/selection_sort_time.csv', sep=';', header=None, names=['n', 'time'])
data4 = pd.read_csv('TimeFiles/shell_sort_time.csv', sep=';', header=None, names=['n', 'time'])

#  
plt.plot(data1['n'], data1['time'], '-', data2['n'], data2['time'], '--', data3['n'], data3['time'], '-.', data4['n'], data4['time'], ':')
plt.xlabel('Number of insertions')
plt.ylabel('Time taken (microsec)')
plt.legend(('bubble sort', 'insertion sort', 'selection sort', 'shell sort'))
plt.title('Quadratic sorts and one cooler:)')
plt.savefig('punct1.png')
plt.show()
```
