```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из входного файла .txt
data1 = pd.read_csv('bubble_sort_time.txt', sep=';', header=None, names=['n', 'time'])
data2 = pd.read_csv('insertion_sort_time.txt', sep=';', header=None, names=['n', 'time'])
data3 = pd.read_csv('selection_sort_time.txt', sep=';', header=None, names=['n', 'time'])
data4 = pd.read_csv('shell_sort_time.txt', sep=';', header=None, names=['n', 'time'])

# Построение графика
plt.plot(data1['n'], data1['time'], '-', data2['n'], data2['time'], '--', data3['n'], data3['time'], '-.', data4['n'], data4['time'], ':')
plt.xlabel('Number of insertions')
plt.ylabel('Time taken (microsec)')
plt.title('Quadratic sorts and one cooler:)')
plt.savefig('punct1.png')
plt.show()
```
