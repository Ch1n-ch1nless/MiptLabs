```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из входного файла .txt
data = pd.read_csv('da_graphic_points.txt', sep=';', header=None, names=['n', 'time'])

# Построение графика
plt.plot(data['n'], data['time'])
plt.xlabel('Number of insertions')
plt.ylabel('Time taken (s)')
plt.title('Insertion Time vs Number of Insertions')
plt.savefig('stack_on_dynamic_array.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из входного файла .txt
data1 = pd.read_csv('da_graphic_points.txt', sep=';', header=None, names=['n', 'time'])
data2 = pd.read_csv('sll_graphic_points.txt', sep=';', header=None, names=['n', 'time'])

# Построение графика
plt.plot(data1['n'], data1['time'], '-', data2['n'], data2['time'], '--')
plt.xlabel('Number of insertions')
plt.ylabel('Time taken (ms)')
plt.title('Insertion Time / Number of Insertions')
plt.savefig('work_time.png')
plt.show()
```
